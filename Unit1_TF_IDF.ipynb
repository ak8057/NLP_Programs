```python
import math
import re
from collections import Counter

# Documents
doc1 = "He is a good boy. She is also good."
doc2 = "Radhika is a good person."

documents = [doc1, doc2]

# -------------------------
# Text preprocessing
# -------------------------
def tokenize(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text.split()

tokenized_docs = [tokenize(doc) for doc in documents]

# -------------------------
# Vocabulary
# -------------------------
vocab = sorted(set(word for doc in tokenized_docs for word in doc))
print("Vocabulary:\n", vocab)

# -------------------------
# TF (Term Frequency)
# -------------------------
tf_list = []
for doc in tokenized_docs:
    word_count = Counter(doc)
    total_words = len(doc)
    tf = {word: word_count.get(word, 0) / total_words for word in vocab}
    tf_list.append(tf)

print("\nTF:")
for i, tf in enumerate(tf_list):
    print(f"Doc {i+1}:", tf)

# -------------------------
# IDF (Inverse Document Frequency)
# -------------------------
N = len(documents)
idf = {}

for word in vocab:
    doc_freq = sum(1 for doc in tokenized_docs if word in doc)
    idf[word] = math.log(N / doc_freq)

print("\nIDF:\n", idf)

# -------------------------
# TF-IDF
# -------------------------
tfidf_list = []
for tf in tf_list:
    tfidf = {word: tf[word] * idf[word] for word in vocab}
    tfidf_list.append(tfidf)

print("\nTF-IDF:")
for i, vec in enumerate(tfidf_list):
    print(f"Doc {i+1}:", vec)
```

```python
import re
from collections import Counter

# -------------------------
# Dataset directly in code
# -------------------------
tweets = [
    "I love AI!!! Visit https://ai.com now 😍",
    "<b>Machine learning</b> is AMAZING!!!",
    "NLP is coool & useful #technology",
    "Radhika likes data science a lot!!!",
    "He is a good boy, she is also good."
]

# -------------------------
# Simple stopword list (manual)
# -------------------------
stop_words = {
    "is","a","the","and","or","in","on","at","to","for","of",
    "he","she","it","also","now"
}

# -------------------------
# Cleaning function
# -------------------------
def clean_text(text):
    text = text.lower()                              # lowercase
    text = re.sub(r'http\S+|www\S+', '', text)       # remove urls
    text = re.sub(r'<.*?>', '', text)                # remove html
    text = re.sub(r'[^a-zA-Z\s]', '', text)          # remove punctuation
    words = text.split()
    words = [w for w in words if w not in stop_words]  # remove stopwords
    return " ".join(words)

cleaned = [clean_text(t) for t in tweets]

# -------------------------
# Very simple stemming (basic)
# -------------------------
def simple_stem(word):
    suffixes = ['ing', 'ly', 'ed', 's']
    for s in suffixes:
        if word.endswith(s):
            return word[:-len(s)]
    return word

stemmed = [" ".join(simple_stem(w) for w in text.split()) for text in cleaned]

# -------------------------
# Show results
# -------------------------
print("Original:\n", tweets)
print("\nCleaned:\n", cleaned)
print("\nStemmed:\n", stemmed)
```
